```python
import pandas as pd 
import numpy as np 

import os
from langchain_community.document_loaders import PyPDFLoader
```

```python
data_path = r"..\..\data"
pdf_files = [os.path.join(root, file) for root, _, files in os.walk(data_path) for file in files if ".pdf" in file]

pdf_pages = []
for file in pdf_files:
    print(file)
    loader = PyPDFLoader(file)
    async for page in loader.alazy_load():
        pdf_pages.append(page)
```

```python
print(f"{pdf_pages[18].metadata}\n")
print(pdf_pages[18].page_content)
pdf_pages[18].page_content
```

```python
pdf_pages[-1]
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)

documents = []
for page in pdf_pages:
    print(page.metadata['source'])
    documents.extend(text_splitter.split_text(page.page_content))
```

```python
documents
```

```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
embeddings = embedding_model.encode(documents, convert_to_tensor=False)
```

```python
from langchain.vectorstores import Chroma
persist_directory = "books"
vector_store = Chroma.from_texts(
    texts=documents,
    embedding=embeddings,
    persist_directory="books",
    collection_name="book_store"
)
vector_store.persist()
# vector_store = Chroma.from_embeddings(embeddings=embeddings, texts=documents, persist_directory=persist_directory)
```
